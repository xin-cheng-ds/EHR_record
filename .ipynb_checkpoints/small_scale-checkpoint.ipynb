```python
sample_demographic_train = demographic_train.sample(frac=0.1)
sample_labs_train = labs_train.sample(frac=0.1)
sample_vitals_train = vitals_train.sample(frac=0.1)
```

```python
#EDA

#remove weight column
def demographic_process(df):
    return df.drop('weight', axis=1)

# remove unit of measure in labs and vitals, the unit of measure shouldn't have any predict power
def labs_process(df):
    return df.drop('Unit of Measure', axis=1)

# It's the same function as labs_process, repeat for conformidity
def vitals_process(df):
    return df.drop('Unit of Measure', axis=1)

```

```python
sample_demographic_train = demographic_process(sample_demographic_train)
sample_labs_train = labs_process(sample_labs_train)
sample_vitals_train = vitals_process(sample_vitals_train)
```

```python


def pivot_data(df, index_col, columns_col, values_col, aggfunc='mean'):
    return df.pivot_table(index=index_col, columns=columns_col, values=values_col, aggfunc=aggfunc).reset_index()

def merge_dataframes(demo_df, vitals_df, labs_df, on_col='patient_id'):
    merged_df = pd.merge(demo_df, vitals_df, on=on_col, how='inner')
    merged_df = pd.merge(merged_df, labs_df, on=on_col, how='inner')
    return merged_df



```

```python
# Pivot the vitals and labs dataframes
vitals_pivoted = pivot_data(vitals, 'patient_id', 'Vital', 'Value')
labs_pivoted = pivot_data(labs, 'patient_id', 'Lab', 'Value')

# Merge the demo, pivoted vitals, and pivoted labs dataframes
final_df = merge_dataframes(demo, vitals_pivoted, labs_pivoted)

# The final dataframe is now stored in 'final_df'
```
